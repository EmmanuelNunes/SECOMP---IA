## Criar Coleção
Coleções no ChromaDB são como 'tabelas' de um banco tradicional, mas armazenam vetores.

```python
import chromadb

client = chromadb.Client()
collection = client.create_collection("secomp")
print("Coleção 'secomp' criada com sucesso!")
```

## Inserir Documento
Vamos inserir nosso primeiro pedaço de contexto (chunk).

```python
collection.add(
    documents=["RAG significa Retrieval Augmented Generation"],
    ids=["doc1"]
)
print("Documento indexado!")
```

## Consulta (Retrieval)
Realizamos uma busca semântica para recuperar a informação mais semelhante.

```python
results = collection.query(
    query_texts=["O que é RAG?"]
)

print("\n--- RESULTADOS DA BUSCA ---")
print(results["documents"])
```
