## Objetivo

## Estrutura Base
A estrutura recomendada para prompts robustos é:
- **ROLE** (Papel do Assistente)
- **TASK** (Tarefa)
- **CONTEXT** (Contexto)
- **CONSTRAINTS** (Restrições)
- **OUTPUT FORMAT** (Formato de Saída)
