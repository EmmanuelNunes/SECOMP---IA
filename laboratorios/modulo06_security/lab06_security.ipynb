**Criar um filtro preventivo.**

```python
def filtro_simples(texto):
    if "ignore todas as instruções" in texto.lower():
        return False # Bloquear
    return True

if not filtro_simples(documento_usuario):
    print("ALERTA: Tentativa de Prompt Injection Bloqueada!")
```
