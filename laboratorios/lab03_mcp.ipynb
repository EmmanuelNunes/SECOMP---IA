Na arquitetura do MCP, componentes são divididos em `Tool`, `Resource` ou `Prompt`.

Onde se enquadra a integração com o **GitHub**?
- *Resposta:* **Tool** (pois permite executar ações como pull/push/commit).

## Exercício
